```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r'C:\Users\SEO01\Downloads\new age and gender data.csv')
```

```python
df
```

```python
total_reach= df['Reach'].sum()
total_impression=df['Impressions'].sum()
total_amount=df['Amount spent (PKR)'].sum()
```

```python
d_summary=pd.DataFrame({
    'Metric':['Reach','Impressions','Amount spent (PKR)'],
    'Value':[total_reach,total_impression,total_amount]
})
```

## Summary Statistics

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Metric', y='Value', data=d_summary)
plt.title('Summary Statistics')
plt.xlabel('Metric')
plt.ylabel('Value')
plt.ticklabel_format(style='plain', axis='y') 
plt.show()
```

```python
data_agg=df.groupby(['Age','Gender']).agg({
    'Reach':'sum',
    'Impressions':'sum',
    'Amount spent (PKR)': 'sum',
    'Link clicks':'sum'
}).reset_index()
```

```python
total_reach
```

```python
total_amount
```

## Reach By Age Group and Gender

```python
plt.figure(figsize=(14,4))
plt.subplot(1,2,1)
sns.barplot(x='Age', y='Reach', hue='Gender', data=data_agg)
plt.title('Reach By Age Group and Gender')
plt.xlabel('Age')
plt.ylabel('Reach')
plt.show()
```

## Impressions by Age Group and Gender

```python
plt.figure(figsize=(14, 4))
plt.subplot(1, 2, 1)
sns.barplot(x='Age', y='Impressions', hue='Gender', data=data_agg)
plt.title('Impressions by Age Group and Gender')
plt.xlabel('Age')
plt.ylabel('Impressions')
plt.ticklabel_format(style='plain', axis='y')
plt.show()
```

```python
data_agg['Cost per Result']= data_agg['Amount spent (PKR)'] / data_agg['Link clicks']
data_agg['CTR']= data_agg['Link clicks'] / data_agg['Impressions']
```

## Cost Per Result by Age Group and Gender

```python
plt.figure(figsize=(14,6))
plt.subplot(1,2,1)
sns.barplot(x='Age', y='Cost per Result', hue = 'Gender', data=data_agg)
plt.title('Cost Per Result by Age Group and Gender')
plt.xlabel('Age')
plt.ylabel('Cost Per Result')
plt.show()
```

## Link Clicks by Gender and Age

```python
plt.figure(figsize=(14,6))
plt.subplot(1,2,1)
sns.barplot(x='Age', y='Link clicks', hue = 'Gender', data=data_agg)
plt.title('Link Clicks by Gender and Age')
plt.xlabel('Age')
plt.ylabel('Link Clicks')
plt.show()
```

## CTR by Age Group and Gender

```python
plt.figure(figsize=(14,6))
plt.subplot(1,2,1)
sns.barplot(x='Age', y='CTR', hue = 'Gender', data=data_agg)
plt.title('CTR by Age Group and Gender')
plt.xlabel('Age')
plt.ylabel('CTR')
plt.show()
```
